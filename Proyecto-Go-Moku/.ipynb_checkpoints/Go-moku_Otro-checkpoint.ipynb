## Problema: Go-moku

Go-moku es un juego de mesa para dos personas. Se juega con piedras blancas y negras en un tablero de Go de intersección de 15x15 o 19x19.

Las negras juegan primero y los jugadores se alternan para colocar una piedra de su color en una intersección vacía. No hay restricción de dónde colocar, los jugadores colocan sus piedras alternativamente hasta que se recolectan una fila ininterrumpida de cinco piedras, ya sea horizontal, vertical o diagonalmente.

<img src="img/hh.jpeg" alt="drawing" width="250"/>

## Objetivo del proyecto

El objetivo es que el computador genere un tablero en el que un jugador gana.

## Representación de las letras proposicionales.

$N(x,y,t)$ es verdadero sii en el turno $t$ hay una piedra negra en la intersección $x$, $y$  
$B(x,y,t)$ es verdadero sii en el turno $t$ hay una piedra blanca en la intersección $x$, $y$.


$x$ son las filas y $y$ son las columnas del tablero.


## Etapa de la codificación de las letras proposicionales

(Basado en el taller de clase de Codificación)

```python
class objetoCodificacionNegras:
    
    def __init__ (self,Nf,Nc,Nn,chrInit) :
        self.Nf = Nf
        self.Nc = Nc
        self.Nn = Nn
        self.chrInit = chrInit

    def codifica(self,f,c,n) :
        v1 = self.Nc * f + c
        return self.Nn * v1 + n

    def decodifica(self,codigo):
        n = codigo % self.Nn
        v1 = int(codigo / self.Nn)
        c = v1 % self.Nc
        f = int(v1 / self.Nc)
        return f, c, n
    
    def N(self,f,c,n):
        codigo = self.codifica(f,c,n)
        return chr(self.chrInit + codigo)
    
    def Ninv(self,codigo) :
        v = ord(codigo)-self.chrInit
        return self.decodifica(v)
```

```python
class objetoCodificacionBlancas:
    
    def __init__ (self,Nf,Nc,Nn,chrInit) :
        self.Nf = Nf
        self.Nc = Nc
        self.Nn = Nn
        self.chrInit = chrInit

    def codifica(self,f,c,n) :
        v1 = self.Nc * f + c
        return self.Nn * v1 + n

    def decodifica(self,codigo):
        n = codigo % self.Nn
        v1 = int(codigo / self.Nn)
        c = v1 % self.Nc
        f = int(v1 / self.Nc)
        return f, c, n
    
    def B(self,f,c,n):
        codigo = self.codifica(f,c,n)
        return chr(self.chrInit + codigo)
    
    def Binv(self,codigo) :
        v = ord(codigo)-self.chrInit
        return self.decodifica(v)
```

## Restricciones del problema

#### Reglas:

1. La primera piedra en ser colocada en el tablero, debe ser una piedra negra, no hay restricción en cual intersección puede ser colocada.

2. No puede haber más de una piedra en cada intersección, sin importar el color que está.

3. Se gana con cinco piedras del mismo color seguidas en línea vertical, horizontal o diagonal.

4. Si hay una piedra en una posición del tablero, esta se queda ahí para todos los turnos siguientes.

5. (Por evaluar) Se coloca una piedra a la vez en cada ronda.


```python
letrasProposicionalesNegras = [chr(x) for x in range(256,1744)]
letrasProposicionalesBlancas = [chr(x) for x in range(2001,3745)]

letrasProposicionales = letrasProposicionalesNegras + letrasProposicionalesBlancas
conectivos = ["Y", "O", ">", "="]
```

```python
class Tree(object):
    def __init__(self, label, left, right):
        self.left = left
        self.right = right
        self.label = label
        
def String2Tree(A):
    # Crea una formula como tree dada una formula como cadena escrita en notacion polaca inversa
    # Input: - A, lista de caracteres con una formula escrita en notacion polaca inversa
    #        - letrasProposicionales, lista de letras proposicionales
    #        - conectivos, lista de conectivos
    # Output: formula como tree
    pila = []
    for c in A:
        # print("Examinando " + str(c))
        if c in letrasProposicionales:
            # print(u"El símbolo es letra proposicional")
            pila.append(Tree(c, None, None))
        elif c == '-':
            # print("Negamos")
            formulaAux = Tree(c, None, pila[-1])
            del pila[-1]
            pila.append(formulaAux)
        elif c in conectivos:
        # print("Unimos mediante conectivo")
            formulaAux = Tree(c, pila[-1], pila[-2])
            del pila[-1]
            del pila[-1]
            pila.append(formulaAux)
        else:
            print(u"Hay un problema: el símbolo " + str(c) + " no se reconoce")
    return pila[-1]

def Inorderp(f):
    # Imprime una formula como cadena dada una formula como arbol
    # de manera que las letras proposicionales sean legibles.
    # Input: tree, que es una formula de logica proposicional
    # Output: string de la formula

    if f.right == None:
        if f.label in letrasProposicionalesNegras:
            f,c,n = cods1.Ninv(f.label)
            return f"N({f},{c},{n})"
        elif f.label in letrasProposicionalesBlancas:
            f,c,n = cods2.Binv(f.label)
            return f"B({f},{c},{n})"
        
    elif f.label == '-':
        return f.label + Inorderp(f.right)
    else:
        return "(" + Inorderp(f.left) + f.label + " " + Inorderp(f.right) + ")"
```

### Pasos para la regla 1: 
La primera piedra en ser colocada en el tablero, debe ser una piedra negra, no hay restricción en cual intersección puede ser colocada.

1.1 \begin{align}
\left( \bigvee_{x = b}^{f} \bigvee_{y = 2}^{6} N(x,y,1) \right)
\end{align}

```python
inicial = True
def negras_comienza():
    inicial = True
    Nf = 6
    Nc = 6
    
    for f in range(1, Nf):
        for c in range(1, Nc):
            if inicial:
                formula = cods1.N(f,c,1)
                inicial = False
            else:
                formula = cods1.N(f,c,1) + formula + "O"
    return formula

formula = negras_comienza()
A = String2Tree(formula)
print(Inorderp(A))
```

1.2 \begin{align}
\left( \bigwedge_{x = b}^{f} \bigwedge_{y = 2}^{6} \neg B(x,y,1) \right)
\end{align}

```python
inicial = True
def no_blancas_comienza():
    inicial = True
    Nf = 6
    Nc = 6
    
    for f in range(1, Nf):
        for c in range(1, Nc):
            if inicial:
                formula = cods2.B(f,c,1) + "-"
                inicial = False
            else:
                formula = cods2.B(f,c,1) + "-" + formula + "Y"
    return formula

formula = no_blancas_comienza()
A = String2Tree(formula)
print(Inorderp(A))
```

1.4 \begin{align}
N(x,y,1) \rightarrow \left(\bigwedge_{(u,v) \neq (x,y)} \neg N(u,v,1)\right)
\end{align}

```python
inicial = True
def no_dif_negras(x,y):
    inicial = True
    Nf = 6
    Nc = 6
    rango = [(u,v) for u in range(1, cods1.Nf) for v in range(1, cods1.Nc)]
    rango = [casilla for casilla in rango if casilla != (x,y)]
    for (u,v) in rango:
        if inicial:
            formula = cods1.N(u,v,1) + '-'
            inicial = False
        else:
            formula = cods1.N(u,v,1) + '-' + formula + "Y"
    
    formula = formula + cods1.N(x,y,1) + '>'
    return formula

formula = no_dif_negras(0,0)
A = String2Tree(formula)
print(Inorderp(A))
```

1.5 \begin{align*}
\left( \bigwedge_{x = b}^{f} \bigwedge_{y = 2}^{6} \left( N(x,y,1) \rightarrow \left(\bigwedge_{(u,v) \neq (x,y)} \neg N(u,v,1)\right) \right) \right)
\end{align*}

```python
def no_dif_negras_():
    inicial = True
    for f in range(1, cods1.Nf):
        for c in range(1, cods1.Nc): 
            if inicial:
                formula = no_dif_negras(f,c)
                inicial = False
            else:
                formula = no_dif_negras(f,c) + formula + "Y"
    return formula
```

1.5 \begin{align*}
    \left( \bigvee_{x = b}^{f} \bigvee_{y = 2}^{6} N(x,y,1) \right) \wedge \left( \bigwedge_{x = b}^{f} \bigwedge_{y = 2}^{6} \neg B(x,y,1) \right)  \wedge \left( \bigwedge_{x = b}^{f} \bigwedge_{y = 2}^{6} \left( N(x,y,1) \rightarrow \left(\bigwedge_{(u,v) \neq (x,y)} \neg N(u,v,1)\right) \right) \right)

```python
def regla1():
    formula1 = negras_comienza()
    formula2 = no_blancas_comienza()
    formula3 = no_dif_negras_()
    formula = formula3 + formula2 + 'Y' + formula1 + 'Y'
    return formula
```

### Pasos para la regla 2: 

\begin{align*}
\left( N(x,y,t) \rightarrow \neg B(x,y,t) \right) \wedge \left( B(x,y,t) \rightarrow \neg N(x,y,t) \right)
\end{align*}

```python
def no_piedras_misma_interseccion(x,y,t):
    formula1 = cods2.B(x,y,t) + "-" + cods1.N(x,y,t) + ">"
    formula2 = cods1.N(x,y,t) + "-" + cods2.B(x,y,t) + ">"
    formula = formula2 + formula1 + "Y"
    return formula

formula = no_piedras_misma_interseccion(1,1,2)
A = String2Tree(formula)
print(Inorderp(A))
```

\begin{align*}
  \bigwedge_{t = 1}^{25} \bigwedge_{x = b}^{f} \bigwedge_{y = 2}^{6} \left( N(x,y,t) \rightarrow \neg B(x,y,t) \right) \wedge \left( B(x,y,t) \rightarrow \neg N(x,y,t) \right)
\end{align*}

```python
def regla2():
    inicial = True
    Nf = 6
    Nc = 6
    Nt = 25
    for f in range(1, Nf):
        for c in range(1, Nc): 
            for t in range(1, Nt+1):
                if inicial:
                    formula = no_piedras_misma_interseccion(f,c,t)
                    inicial = False
                else:
                    formula = no_piedras_misma_interseccion(f,c,t) + formula + 'Y'
    return formula       
```

## Visualización de una solución


```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.offsetbox import AnnotationBbox, OffsetImage
from copy import deepcopy
%matplotlib inline
```

```python
def tablero(f):

    #Dibuja las cuadriculas coordenadas dada una formula f
    #Input: 
        # - f : es un diccionario con literales y sus valores de verdad.
        # -color: color de la cuadricula.
    #Output:
        # archivo de imagen: gomoku.jpg
    
    # Dibujo el tablero
    fig, axes= plt.subplots()
    plt.yticks(np.arange(0, 1, 1./6))
    plt.xticks([0, 1, 2, 3, 4, 5, 6], ['a', 'b', 'c', 'd', 'e', 'f', 'g'])
    
    axes.get_xaxis().set_visible(True)
    axes.get_yaxis().set_visible(True)
    
    step = 1./6
    triangulos = []
    

    # Creo las líneas del tablero
    for j in range(14): #Determina que sea un tablero 6x6
        locacion = j * step
        # Crea linea horizontal en el rectangulo
        triangulos.append(patches.Rectangle(*[(0, step + locacion), 1, 0.005],\
                facecolor='black', color="e28f4d"))
        # Crea linea vertical en el rectangulo
        triangulos.append(patches.Rectangle(*[(step + locacion, 0), 0.005, 1],\
                facecolor='black', color="e28f4d"))
    
    # crea el tablero
    for t in triangulos:
        axes.add_patch(t)
    
    # Creando las direcciones en la imagen de acuerdo a literal
    direcciones = {}
    direcciones[0] = [0.166, 0.5]
    direcciones[1] = [0.33, 0.31]
    direcciones[2] = [0.33, 0.5]
    direcciones[3] = [0.155, 0.7]
    direcciones[4] = [0.5, 0.5]
    direcciones[5] = [0.5, 0.8]
    direcciones[6] = [0.68, 0.5]
    direcciones[7] = [0.155, 0.88]
    direcciones[8] = [0.84, 0.5]
    
    for i in f:
        
        if i%2 == 0:#Determina la paridad de la llave
        
        #Teniendo en cuenta las reglas del juego, se sabe que la primera pieza en colocarse debe ser negra.
            if f[i] != 0: #Si el valor es diferente de Falso, se utilizara la 
            #imagen de la pieza negra.
                p_negra = plt.imread("img/p_negra.png", format='png')
                imagebox = OffsetImage(p_negra, zoom=0.1)
                imagebox.image.axes = axes
                ab = AnnotationBbox(imagebox, direcciones[int(i)], frameon=False)
                axes.add_artist(ab)
        
        #Situación analoga con las fichas blancas.
        if i%2 != 0:
            if f[i] != 0:
                p_blanca = plt.imread("img/p_blanca.png", format= "png")
                image2box = OffsetImage(p_blanca, zoom = 0.1)
                image2box.image.axes = axes
                ab = AnnotationBbox(image2box, direcciones[int(i)], frameon=False)
                axes.add_artist(ab)
    
    #plt.savefig("gomoku" + ".png")
    plt.show()

    
```
