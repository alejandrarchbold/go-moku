## Problema: Go-moku

Go-moku es un juego de mesa para dos personas. Se juega con piedras blancas y negras en un tablero de Go de intersección de 15x15 o 19x19.

Las negras juegan primero y los jugadores se alternan para colocar una piedra de su color en una intersección vacía. No hay restricción de dónde colocar, los jugadores colocan sus piedras alternativamente hasta que se recolectan una fila ininterrumpida de cinco piedras, ya sea horizontal, vertical o diagonalmente.

<img src="img/hh.jpeg" alt="drawing" width="250"/>

## Objetivo del proyecto

El objetivo es que el computador genere un tablero en el que un jugador gana.

## Representación de las letras proposicionales.

$N(x,y,t)$ es verdadero sii en el turno $t$ hay una piedra negra en la intersección $x$, $y$  
$B(x,y,t)$ es verdadero sii en el turno $t$ hay una piedra blanca en la intersección $x$, $y$.


$x$ son las filas y $y$ son las columnas del tablero.


## Etapa de la codificación de las letras proposicionales

(Basado en el taller de clase de Codificación)

```python
class objetoCodificacionNegras:
    
    def __init__ (self,Nf,Nc,Nn,chrInit) :
        self.Nf = Nf
        self.Nc = Nc
        self.Nn = Nn
        self.chrInit = chrInit

    def codifica(self,f,c,n) :
        v1 = self.Nc * f + c
        return self.Nn * v1 + n

    def decodifica(self,codigo):
        n = codigo % self.Nn
        v1 = int(codigo / self.Nn)
        c = v1 % self.Nc
        f = int(v1 / self.Nc)
        return f, c, n
    
    def N(self,f,c,n):
        codigo = self.codifica(f,c,n)
        return chr(self.chrInit + codigo)
    
    def Ninv(self,codigo) :
        v = ord(codigo)-self.chrInit
        return self.decodifica(v)
```

```python
class objetoCodificacionBlancas:
    
    def __init__ (self,Nf,Nc,Nn,chrInit) :
        self.Nf = Nf
        self.Nc = Nc
        self.Nn = Nn
        self.chrInit = chrInit

    def codifica(self,f,c,n) :
        v1 = self.Nc * f + c
        return self.Nn * v1 + n

    def decodifica(self,codigo):
        n = codigo % self.Nn
        v1 = int(codigo / self.Nn)
        c = v1 % self.Nc
        f = int(v1 / self.Nc)
        return f, c, n
    
    def B(self,f,c,n):
        codigo = self.codifica(f,c,n)
        return chr(self.chrInit + codigo)
    
    def Binv(self,codigo) :
        v = ord(codigo)-self.chrInit
        return self.decodifica(v)
```

#### Dimensiones de la información para la visualización de las Negras

```python
# Se define el rango de opciones para cada dimensión de información
Nfilas = 6
Ncolumnas = 6
Nnumeros = 37

# Se crea el objeto de codificación
cods1 = objetoCodificacionNegras(Nfilas,Ncolumnas,Nnumeros,256)

# Se recorren todas las opciones de las dimensiones de información para visualizar su codificación
letras_negras = []
for n in range(1, Nnumeros):
    print("Turno: "+str(n))
    print("filas x columnas")
    for f in range(Nfilas):
        for c in range(Ncolumnas):
            cod = cods1.N(f,c,n)
            print(cod,end=" ")
            letras_negras.append(cod)
        print("")
    print('\n')
```

## Restricciones del problema

#### Reglas:

1. La primera piedra en ser colocada en el tablero, debe ser una piedra negra, no hay restricción en cual intersección puede ser colocada.

2. No puede haber más de una piedra en cada intersección, sin importar el color que está.

3. Se gana con cinco piedras del mismo color seguidas en línea vertical, horizontal o diagonal.

4. Si hay una piedra en una posición del tablero, esta se queda ahí para todos los turnos siguientes.

5. (Por evaluar) Se coloca una piedra a la vez en cada ronda.


```python
letrasProposicionalesNegras = [chr(x) for x in range(256, 1744)]
letrasProposicionalesBlancas = [chr(x) for x in range(2001,3745)]

letrasProposicionales = letrasProposicionalesNegras + letrasProposicionalesBlancas
conectivos = ["Y", "O", ">", "="]
```

```python
class Tree(object):
    def __init__(self, label, left, right):
        self.left = left
        self.right = right
        self.label = label
        
def String2Tree(A):
    # Crea una formula como tree dada una formula como cadena escrita en notacion polaca inversa
    # Input: - A, lista de caracteres con una formula escrita en notacion polaca inversa
    #        - letrasProposicionales, lista de letras proposicionales
    #        - conectivos, lista de conectivos
    # Output: formula como tree
    pila = []
    for c in A:
        # print("Examinando " + str(c))
        if c in letrasProposicionales:
            # print(u"El símbolo es letra proposicional")
            pila.append(Tree(c, None, None))
        elif c == '-':
            # print("Negamos")
            formulaAux = Tree(c, None, pila[-1])
            del pila[-1]
            pila.append(formulaAux)
        elif c in conectivos:
        # print("Unimos mediante conectivo")
            formulaAux = Tree(c, pila[-1], pila[-2])
            del pila[-1]
            del pila[-1]
            pila.append(formulaAux)
        else:
            print(u"Hay un problema: el símbolo " + str(c) + " no se reconoce")
    return pila[-1]

def Inorderp(f):
    # Imprime una formula como cadena dada una formula como arbol
    # de manera que las letras proposicionales sean legibles.
    # Input: tree, que es una formula de logica proposicional
    # Output: string de la formula

    if f.right == None:
        if f.label in letrasProposicionalesNegras:
            f,c,n = cods1.Ninv(f.label)
            return f"N({f},{c},{n})"
        elif f.label in letrasProposicionalesBlancas:
            f,c,n = cods2.Binv(f.label)
            return f"B({f},{c},{n})"
        
    elif f.label == '-':
        return f.label + Inorderp(f.right)
    else:
        return "(" + Inorderp(f.left) + f.label + " " + Inorderp(f.right) + ")"
```

### Pasos para la regla 1: 

1.1 Hay una negra en alguna coordenada ... \begin{align}
\left( \bigvee_{x = b}^{f} \bigvee_{y = 2}^{6} N(x,y,1) \right)
\end{align}

```python
inicial = True
def negras_comienza():
    inicial = True
    Nf = 6
    Nc = 6
    
    for f in range(1, Nf):
        for c in range(1, Nc):
            if inicial:
                formula = cods1.N(f,c,1)
                inicial = False
            else:
                formula = cods1.N(f,c,1) + formula + "O"
    return formula

formula = negras_comienza()
A = String2Tree(formula)
print(Inorderp(A))
```

1.2 No puede haber una blanca en el primer turno ... 

\begin{align}
\left( \bigwedge_{x = b}^{f} \bigwedge_{y = 2}^{6} \neg B(x,y,1) \right)
\end{align}

```python
inicial = True
def no_blancas_comienza():
    inicial = True
    Nf = 6
    Nc = 6
    
    for f in range(1, Nf):
        for c in range(1, Nc):
            if inicial:
                formula = cods2.B(f,c,1) + "-"
                inicial = False
            else:
                formula = cods2.B(f,c,1) + "-" + formula + "Y"
    return formula

formula = no_blancas_comienza()
A = String2Tree(formula)
print(Inorderp(A))
```

1.3 Si hay una negra en una coordenada, no puede haber otra negra en otra coordenada...  

\begin{align}
N(x,y,1) \rightarrow \left(\bigwedge_{(u,v) \neq (x,y)} \neg N(u,v,1)\right)
\end{align}

```python
inicial = True
def no_dif_negras(x,y):
    inicial = True
    Nf = 6
    Nc = 6
    rango = [(u,v) for u in range(1, cods1.Nf) for v in range(1, cods1.Nc)]
    rango = [casilla for casilla in rango if casilla != (x,y)]
    for (u,v) in rango:
        if inicial:
            formula = cods1.N(u,v,1) + '-'
            inicial = False
        else:
            formula = cods1.N(u,v,1) + '-' + formula + "Y"
    
    formula = formula + cods1.N(x,y,1) + '>'
    return formula

formula = no_dif_negras(0,0)
A = String2Tree(formula)
print(Inorderp(A))
```

1.4  Para todas las posibilidades del paso 1.3

\begin{align*}
\left( \bigwedge_{x = b}^{f} \bigwedge_{y = 2}^{6} \left( N(x,y,1) \rightarrow \left(\bigwedge_{(u,v) \neq (x,y)} \neg N(u,v,1)\right) \right) \right)
\end{align*}

```python
def no_dif_negras_():
    inicial = True
    for f in range(1, cods1.Nf):
        for c in range(1, cods1.Nc): 
            if inicial:
                formula = no_dif_negras(f,c)
                inicial = False
            else:
                formula = no_dif_negras(f,c) + formula + "Y"
    return formula
```

```python
formula = no_dif_negras_()
A = String2Tree(formula)
print(Inorderp(A))
```

1.5 Finalmente, uniendo las funciones anteriores, obtenemos la **regla 1**: 

\begin{align*}
    \left( \bigvee_{x = b}^{f} \bigvee_{y = 2}^{6} N(x,y,1) \right) \wedge \left( \bigwedge_{x = b}^{f} \bigwedge_{y = 2}^{6} \neg B(x,y,1) \right)  \wedge \left( \bigwedge_{x = b}^{f} \bigwedge_{y = 2}^{6} \left( N(x,y,1) \rightarrow \left(\bigwedge_{(u,v) \neq (x,y)} \neg N(u,v,1)\right) \right) \right)
\end{align*}

La primera piedra en ser colocada en el tablero, debe ser una piedra negra, no hay restricción en cual intersección puede ser colocada.

```python
def regla1():
    formula1 = negras_comienza()
    formula2 = no_blancas_comienza()
    formula3 = no_dif_negras_()
    formula = formula3 + formula2 + 'Y' + formula1 + 'Y'
    return formula
```

```python
formula = regla1()
A = String2Tree(formula)
print(Inorderp(A))
```

### Pasos para la regla 2: 

2.1 No puede haber una misma piedra en una interseccion escogida.

\begin{align*}
\left( N(x,y,t) \rightarrow \neg B(x,y,t) \right) \wedge \left( B(x,y,t) \rightarrow \neg N(x,y,t) \right)
\end{align*}

```python
def no_piedras_misma_interseccion(x,y,t):
    formula1 = cods2.B(x,y,t) + "-" + cods1.N(x,y,t) + ">"
    formula2 = cods1.N(x,y,t) + "-" + cods2.B(x,y,t) + ">"
    formula = formula2 + formula1 + "Y"
    return formula

formula = no_piedras_misma_interseccion(1,1,2)
A = String2Tree(formula)
print(Inorderp(A))
```

2.2. Para todas las posibilidades del paso 2.1. Obtenemos la **regla 2**:

\begin{align*}
  \bigwedge_{t = 1}^{25} \bigwedge_{x = b}^{f} \bigwedge_{y = 2}^{6} \left( N(x,y,t) \rightarrow \neg B(x,y,t) \right) \wedge \left( B(x,y,t) \rightarrow \neg N(x,y,t) \right)
\end{align*}

No puede haber más de una piedra en cada intersección, sin importar el color que está.

```python
def regla2():
    inicial = True
    Nf = 6
    Nc = 6
    Nt = 25
    for f in range(1, Nf):
        for c in range(1, Nc): 
            for t in range(1, Nt+1):
                if inicial:
                    formula = no_piedras_misma_interseccion(f,c,t)
                    inicial = False
                else:
                    formula = no_piedras_misma_interseccion(f,c,t) + formula + 'Y'
    return formula       
```

```python
formula = regla2()
A = String2Tree(formula)
print(Inorderp(A))
```

### Regla 3: 

La **regla 3** es la union de todas las posibles victorias con una 'O':

\begin{align*}
    R3\_D1 \vee R3\_D2 \vee R3 \_ V \vee R3 \_ H
\end{align*}

Se gana con cinco piedras del mismo color seguidas en línea vertical, horizontal o diagonal.

```python
def regla3():
    diagonal_1 = victoria_diagonal_1()
    diagonal_2 = victoria_diagonal_2()
    vertical = victoria_vertical()
    horizontal = victoria_horizontal()
    victoria = horizontal + vertical + 'O' + diagonal_2 + 'O' + diagonal_1 + 'O'
    return victoria
```

### Pasos para la regla 4:

4.1. La piedra que se coloque, sea negra o blanca, se debe quedar ahí para los siguientes turnos.

\begin{align*}
     \left( N(x,y,t) \rightarrow N(x,y,t')\right) \vee \left(B(x,y,t) \rightarrow B(x,y,t')\right)
\end{align*}
    

```python
def piedas_se_quedan(x,y,t1,t2):
    formula1 = cods1.N(x,y,t2) + cods1.N(x,y,t1) + ">"
    formula2 = cods2.B(x,y,t2) + cods2.B(x,y,t1) + ">"
    formula = formula2 + formula1 + "O"
    return formula

formula = piedas_se_quedan(1,1,1,2)
A = String2Tree(formula)
print(Inorderp(A))
```

4.2 La **regla 4** se conforma para todas las posibilidades del paso 4.1

\begin{align*}
     \bigwedge_{t' = t+1}^{10} \bigwedge_{t = 1}^{10} \bigwedge_{x = b}^{f} \bigwedge_{y = 2}^{6} \left( N(x,y,t) \rightarrow N(x,y,t')\right) \vee \left(B(x,y,t) \rightarrow B(x,y,t')\right)
\end{align*}

Si hay una piedra en una posición del tablero, esta se queda ahí para todos los turnos siguientes.

```python
def regla4():
    inicial = True
    Nf = 6
    Nc = 6
    Nt = 10
    for f in range(1, Nf):
        for c in range(1, Nc): 
            for t1 in range(1, Nt):
                for t2 in range(t1+1, Nt):
                    if inicial:
                        formula = piedas_se_quedan(f,c,t1,t2)
                        inicial = False
                    else:
                        formula = piedas_se_quedan(f,c,t1,t2) + formula + 'O'
    return formula  
```

```python
formula = regla4()
A = String2Tree(formula)
print(Inorderp(A))
```

### (Por evaluar) regla 5:

\begin{align*}

Se coloca una piedra a la vez en cada ronda.

```python
def tablero(f):

    #Dibuja las cuadriculas coordenadas dada una formula f
    #Input: 
        # - f : es un diccionario con literales y sus valores de verdad.
    #Output:
        # tablero.
    
    # Dibujo el tablero
    fig, axes= plt.subplots()
    plt.xticks(np.arange(0, 1, 0.165), ["a","b","c","d","e","f","g"])
    plt.yticks(np.arange(0, 1, 0.165), range(1,8))
    
    axes.get_xaxis().set_visible(True)
    axes.get_yaxis().set_visible(True)
    
    axes.set_facecolor('peachpuff') # color del fondo
    
    step = 1./6
    cuadrados = []
    

    # Creo las líneas del tablero
    for j in range(14): #Determina que sea un tablero 6x6, con lineas 7x7
        locacion = j * step
        # Crea linea horizontal en el rectangulo
        cuadrados.append(patches.Rectangle(*[(0, step + locacion), 1, 0.005],\
                facecolor='black'))
        # Crea linea vertical en el rectangulo
        cuadrados.append(patches.Rectangle(*[(step + locacion, 0), 0.005, 1],\
                facecolor='black'))
    
    # crea el tablero
    for t in cuadrados:
        axes.add_patch(t)
    
    for i in f.keys():
        
        if f[i] != 0: #el valor de verdad es diferente de 0.
            fila, columna, turno = cods1.Ninv(i)
            fila2, columna2, turno2 = cods2.Binv(i)
            
        if turno%2 != 0: #si es el turno de las negras, ubicar en una respetiva coordenada.
            p_negra = plt.imread("img/p_negra.png", format='png')
            imagebox = OffsetImage(p_negra, zoom=0.1)
            imagebox.image.axes = axes
            ab = AnnotationBbox(imagebox, [fila*1./6, columna*1./6], frameon=False)
            axes.add_artist(ab)
 
        if turno2%2 == 0: #si es el turno de las blancas, ubicar en una respetiva coordenada.
            p_blanca = plt.imread("img/p_blanca.png", format= "png")
            image2box = OffsetImage(p_blanca, zoom = 0.1)
            image2box.image.axes = axes
            ab = AnnotationBbox(image2box, [fila2*1./6, columna2*1./6], frameon=False)
            axes.add_artist(ab)
    
    plt.show()
    
```

## Algoritmo de transformacion de Tseitin

```python
# Subrutinas para la transformacion de una
# formula a su forma clausal

def enFNC(A):
    # Subrutina de Tseitin para encontrar la FNC de
    # la formula en la pila
    # Input: A (cadena) de la forma
    #                   p=-q
    #                   p=(qYr)
    #                   p=(qOr)
    #                   p=(q>r)
    # Output: B (cadena), equivalente en FNC
    assert(len(A)==4 or len(A)==7), u"Fórmula incorrecta!"
    B = ''
    p = A[0]
    # print('p', p)
    if "-" in A:
        q = A[-1]
        # print('q', q)
        B = "-"+p+"O-"+q+"Y"+p+"O"+q
    elif "Y" in A:
        q = A[3]
        # print('q', q)
        r = A[5]
        # print('r', r)
        B = q+"O-"+p+"Y"+r+"O-"+p+"Y-"+q+"O-"+r+"O"+p
    elif "O" in A:
        q = A[3]
        # print('q', q)
        r = A[5]
        # print('r', r)
        B = "-"+q+"O"+p+"Y-"+r+"O"+p+"Y"+q+"O"+r+"O-"+p
    elif ">" in A:
        q = A[3]
        # print('q', q)
        r = A[5]
        # print('r', r)
        B = q+"O"+p+"Y-"+r+"O"+p+"Y-"+q+"O"+r+"O-"+p
    elif "=" in A:
        q = A[3]
        # print('q', q)
        r = A[5]
        # print('r', r)
        B = q+"O"+"-"+r+"O"+"-"+p+"Y"+"-"+q+"O"+r+"O"+"-"+p+"Y"+"-"+q+"O"+"-"+r+"O"+p+"Y"+q+"O"+r+"O"+p
    else:
        print(u'Error enENC(): Fórmula incorrecta!')

    return B

# Algoritmo de transformacion de Tseitin
# Input: A (cadena) en notacion inorder
# Output: B (cadena), Tseitin
def Tseitin(A, letrasProposicionalesNegrasBlancas):
    
    letrasProposicionalesOtro = [chr(x) for x in range(3746,7290)]
    
    assert(not bool(set(letrasProposicionalesNegrasBlancas) & set(letrasProposicionalesOtro))), u"¡Hay letras proposicionales en común!"

    letrasProposicionales = letrasProposicionalesNegrasBlancas + letrasProposicionalesOtro

    l = [] #lista de conjuciones
    pila = [] #pila
    i = -1  #contador de variables nuevas
    s = A[0] #simbolo de trabajp

    while(len(A) > 0):
        if ((s in letrasProposicionales) and (len(pila) != 0) and (pila[-1] == '-')):
            i -=- 1
            atomo = letrasProposicionalesOtro[i]
            pila = pila[:-1]
            pila.append(atomo)
            l.append(atomo + "=" + "-" + s)
            A = A[1:]

            if (len(A) > 0):
                s = A[0]

        elif (s == ')'):
            w = pila[-1]
            u = pila[-2]
            v = pila[-3]
            pila = pila[:len(pila)-4]
            i -=- 1
            atomo = letrasProposicionalesOtro[i]
            l.append(atomo + "=" + '(' + v + u + w + ')')
            s = atomo

        else:
            pila.append(s)
            A = A[1:]

            if (len(A) > 0):
                s = A[0]

    B= ''

    if (i < 0):
        atomo = pila[-1]

    else:
        atomo = letrasProposicionalesOtro[i]

    for x in l:
        y = enFNC(x)
        B += "Y" + y

    B = atomo + B

    return B

# Subrutina Clausula para obtener lista de literales
# Input: C (cadena) una clausula
# Output: L (lista), lista de literales
# Se asume que cada literal es un solo caracter
def Clausula(C):

    #  IMPLEMENTAR AQUI ALGORITMO CLAUSULA
    L = [] #lista de literales
    while(len(C) > 0):
        s = C[0]

        if (s == '-'):
            L.append(s + C[1])
            C = C[3:]

        else:
            L.append(s)
            C = C[2:]

    return L


# Algoritmo para obtencion de forma clausal
# Input: A (cadena) en notacion inorder en FNC
# Output: L (lista), lista de listas de literales
def formaClausal(A):

    #  IMPLEMENTAR AQUI ALGORITMO FORMA CLAUSAL
    l = []
    i = 0

    while(len(A) > 0):

        if i >= len(A):
                l.append(Clausula(A))
                A = []
        else:
            if A[i] == 'Y':
                l.append(Clausula(A[:i]))
                A = A[i+1:]
                i = 0
            else:
                i += 1

    return l
```

```python
def complemento(letra):
    if len(letra) == 1:
        complemento = '-' + letra
        return complemento

    elif (len(letra) > 1):
        if (letra[0] == '-'):
            return letra[1]

    else:
        raise(f"error, literal invalido {letra}")

def hay_unidad(S):

    for x in S:
        if len(x) == 1:
            return x[0]
    return None

def unit_propagate(S, I):

    unidad = hay_unidad(S)
    while ([] not in S) and (unidad != None):
        Comple = complemento(unidad)
        S = [c for c in S if unidad not in c]
        count = 0
        for c in S:
            if Comple in c:
                c = [x for x in c if x != Comple]
                S[count] = c
            count += 1

        if '-' in unidad:
            I[unidad] = 0
        else:
            I[unidad] = 1
        unidad = hay_unidad(S)
    return S, I

def DPLL(S,I):
    S, I = unit_propagate(S,I)

    if [] in S:
        return "Insatisfacible", {}
    elif len(S) == 0:
        return "Satisfacible", I

    else:

        l = S[0][0]
        S2 = [x for x in S if l not in x]
        Comple = complemento(l)
        count = 0
        for c in S2:
            if Comple in c:
                c = [x for x in c if x != Comple]
                S2[count] = c
            count += 1
        count = 0
        # print("S2 = ", S2)
        I2 = I

        if len(l)==1:
            I2[l] = 1
        else:
            I2[l[1]] = 0

        conj2, interp2 = DPLL(S2,I2)

        if (conj2 == "Satisfacible") and (interp2 == I2):
            return conj2, interp2


        for x in S:
            for p in x:
                if p not in I.keys():
                    l = p
                    break
        Comple = complemento(l)
        S3 = [x for x in S if Comple not in x]
        count = 0

        for c in S3:
            if l in c:
                c = [x for x in c if x != l]
                S3[count] = c
            count += 1
        count = 0

        I3 = I
        if len(l)==1:
            I3[l] = 0
        else:
            I3[l[1]] = 1
        return DPLL(S3, I3)
```

```python
def grafica_final(f):

    #Dibuja las cuadriculas coordenadas dada una formula f
    #Input: 
        # - f : diccionario con la interpretacion final para cada literal de cada turno y coordenada.
    #Output:
        # tablero.
    
    # Dibujo el tablero
    fig, axes= plt.subplots()
    plt.xticks(np.arange(0, 1, 0.165), ["a","b","c","d","e","f","g"])
    plt.yticks(np.arange(0, 1, 0.165), range(1,8))
    
    axes.get_xaxis().set_visible(True)
    axes.get_yaxis().set_visible(True)
    
    axes.set_facecolor('peachpuff') # color del fondo
    
    step = 1./6
    cuadrados = []
    

    # Creo las líneas del tablero
    for j in range(14): #Determina que sea un tablero 6x6, con lineas 7x7
        locacion = j * step
        # Crea linea horizontal en el rectangulo
        cuadrados.append(patches.Rectangle(*[(0, step + locacion), 1, 0.005],\
                facecolor='black'))
        # Crea linea vertical en el rectangulo
        cuadrados.append(patches.Rectangle(*[(step + locacion, 0), 0.005, 1],\
                facecolor='black'))
    
    # crea el tablero
    for t in cuadrados:
        axes.add_patch(t)
    #===================================================================================
    # busquemos el sub-dicionario de f con la situacion final
    t_max = 1
    dic_t_max = dict()
    # Busqueda lineal del turno maximo
    for j in f.keys():
        fila, columna, turno = cods1.Ninv(j)
        fila2, columna2, turno2 = cods2.Binv(j)        
        if turno > t_max:
            t_max = turno
        elif turno2 > t_max:
            t_max = turno2
    # Se agrega todos los literales en el turno maximo a dict_t_max
    for k in f.keys():
        fila, columna, turno = cods1.Ninv(k)
        fila2, columna2, turno2 = cods2.Binv(k)
        if turno == t_max:
            dic_T_max[k] = f[k]
        elif turno2 == t_max:
            dic_t_max[k] = f[k]
    
    # Se grafica la situacion de cada literal segun el turno final    
    for i in dic_t_max.keys():
        
        if dic_t_max[i] != 0: #el valor de verdad es diferente de 0.
            fila, columna, turno = cods1.Ninv(i)
            fila2, columna2, turno2 = cods2.Binv(i)
            
        if turno%2 != 0: #si es el turno de las negras, ubicar en una respetiva coordenada.
            p_negra = plt.imread("img/p_negra.png", format='png')
            imagebox = OffsetImage(p_negra, zoom=0.1)
            imagebox.image.axes = axes
            ab = AnnotationBbox(imagebox, [fila*1./6, columna*1./6], frameon=False)
            axes.add_artist(ab)
 
        if turno2%2 == 0: #si es el turno de las blancas, ubicar en una respetiva coordenada.
            p_blanca = plt.imread("img/p_blanca.png", format= "png")
            image2box = OffsetImage(p_blanca, zoom = 0.1)
            image2box.image.axes = axes
            ab = AnnotationBbox(image2box, [fila2*1./6, columna2*1./6], frameon=False)
            axes.add_artist(ab)
    
    plt.show()
```
